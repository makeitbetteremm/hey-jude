```python
#将unix时间戳转化为本地时间
import pandas as pd
import time
data=pd.read_csv('C:/Users/HP/Desktop/XX/zy/0.csv',index_col=0)

#定义时间转化函数
def Loct(data,i):
    format = '%Y-%m-%d %H:%M'#定义时间格式
    for a in data.index:
        value=time.localtime(data.iloc[a,i])#将unix时间戳转化本地时间
        data.iloc[a,i]=time.strftime(format,value)#将time中的时间转化为上面定义的格式
    return data
#运行函数
for a in range(2):
    Loct(data,a)
outpath1='C:/Users/HP/Desktop/XX/zy/1.csv'
data.to_csv(outpath1)#存储数据
```

```python
#将火星坐标转化为WGS84
#转载自https://www.cnblogs.com/shadrach/p/7640717.html
import pandas as pd
import numpy as np
data=pd.read_csv('C:/Users/HP/Desktop/XX/zy/1.csv',index_col=0)#读取数据
#定义经纬转化函数
def jingwei(data,i,j):
    a = 6378245.0 #克拉索夫斯基椭球参数长半轴a
    ee = 0.00669342162296594323 #克拉索夫斯基椭球参数第一偏心率平方
    PI=np.pi
    lon=data[i]#经度
    lat=data[j]#纬度
    x=lon-105
    y=lat-35
#经度偏差计算
    dLon = 300.0 + x + 2.0 * y + 0.1 * x * x + 0.1 * x * y + 0.1 * np.sqrt(abs(x))
    dLon += (20.0 * np.sin(6.0 * x * PI) + 20.0 * np.sin(2.0 * x * PI)) * 2.0 / 3.0
    dLon += (20.0 * np.sin(x * PI) + 40.0 * np.sin(x / 3.0 * PI)) * 2.0 / 3.0
    dLon += (150.0 * np.sin(x / 12.0 * PI) + 300.0 * np.sin(x / 30.0 * PI)) * 2.0 / 3.0
#纬度偏差计算
    dLat = -100.0 + 2.0 * x + 3.0 * y + 0.2 * y * y + 0.1 * x * y + 0.2 * np.sqrt(abs(x))
    dlat=(20.0 * np.sin(6.0 * x * PI) + 20.0 * np.sin(2.0 * x * PI)) * 2.0 / 3.0
    dLat += (20.0 * np.sin(y * PI) + 40.0 * np.sin(y / 3.0 * PI)) * 2.0 / 3.0
    dLat += (160.0 * np.sin(y / 12.0 * PI) + 320 * np.sin(y * PI / 30.0)) * 2.0 / 3.0
    radLat = lat / 180.0 * PI
    magic = np.sin(radLat)
    magic = 1 - ee * magic * magic
    sqrtMagic =np.sqrt(magic)
    dLat = (dLat * 180.0) / ((a * (1 - ee)) / (magic * sqrtMagic) * PI);
    dLon = (dLon * 180.0) / (a / sqrtMagic * np.cos(radLat) * PI);
    data[i] = lon - dLon
    data[j]= lat - dLat
    return data
jingwei(data,'2','3')
jingwei(data,'4','5')
data.to_csv('C:/Users/HP/Desktop/XX/zy/2.csv')#存储
```

```python
#每个订单行驶的欧几里得距离计算
import pandas as pd
from numpy import sin,arccos,cos,sqrt,pi,arcsin
R=6371.393#由于只限于程度地区，所以认为地球为正球体，取地球的平均半径为正球体半径
data=pd.read_csv('C:/Users/HP/Desktop/XX/zy/2.csv',index_col=0)#读取wgps坐标
lon1=data['2']*pi/180#转化为弧度
lon2=data['4']*pi/180
lat1=data['3']*pi/180
lat2=data['5']*pi/180
#计算经纬差
dlon=lon1-lon2
dlat=lat2-lat1
#以下根据公式进行计算
a=sin(dlat/2)**2+cos(lat1)*cos(lat2)*sin(dlon/2)**2
c=2*arcsin(sqrt(a))
d=c*R
d=pd.DataFrame(d,columns=['0'])#将d转化为DataFrame格式，便于存储
d.to_csv('C:/Users/HP/Desktop/XX/zy/distance.csv')
```

```python
#统计各时段订单频数频率，作图,制表
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.sans-serif']=['SimHei']#作图字体为黑体
plt.rcParams['axes.unicode_minus']=False#负数显示为负号

data=pd.read_csv('C:/Users/HP/Desktop/XX/zy/2.csv',index_col=0)#读取数据
#从订单起始时间提取出接单时段
hour=data['0'].str.split(' ',expand=True)[1].str.split(':',expand=True)[0]#以‘ ’为分隔符，将日期与时间分割，/
#再以分隔符‘：’将小时与分钟分割 ,取小时
hour.to_csv('C:/Users/HP/Desktop/XX/zy/hour.csv')

pinshu=hour.value_counts()#计算各时段订单数
pinshu=pinshu.sort_index()#将数据索引排序，即按时间排序
sum=len(hour)#计算订单总数
pinlv=pinshu/sum#计算频数
#做频率分布曲线图
plt.figure(figsize=(10,6))#设置图片尺寸
plt.plot(pinlv.index,pinlv.values,'bo-.')#画图，蓝色圆标记短点线
plt.xticks(list(range(0,24)),list(range(0,24)))#x轴显示各个时段
plt.xlabel('时段（h）')#命名x轴
plt.tight_layout()#利用自带函数调整布局
plt.savefig('C:/Users/HP/Desktop/XX/zy/频率分布曲线图.png')
#制表
p=pd.DataFrame(pinshu.values*10,columns=['频数'],index=list(range(0,24)))#转化为Dataframe格式，由于采用系统抽样所以*10以接近真实值
p['频率f']=pinlv.values
p['频率%']=p['频率f'].map(lambda x:'%.2f%%'%(x*100))#将小数转化为百分数
p.to_csv('C:/Users/HP/Desktop/XX/zy/pin.csv')#存储
```

```python
#计算出租车各时段行驶平均速度
import pandas as pd
import matplotlib.pyplot as plt
hour=pd.read_csv('C:/Users/HP/Desktop/XX/zy/hour.csv',index_col=0,header=None)
time=pd.read_csv('C:/Users/HP/Desktop/XX/zy/time.csv',index_col=0)
distance=pd.read_csv('C:/Users/HP/Desktop/XX/zy/distance.csv',index_col=0)
v=pd.DataFrame(distance/time*60)#计算速度

v['1']=hour.iloc[:,0]#定义v的第二列为时段值

u=v['0'].groupby(v['1']).mean()#计算各时段速度平均值

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(10,6))
plt.plot(u.index,u.values,'.-')#画折线图
plt.xticks(list(range(0,24)),list(range(0,24)))#x轴显示24小时全部时段
for i in range(24):
    plt.annotate(round(u[i],2),xy=(i,u[i]),xytext=(i-0.1,u[i]+0.5))#添加注释
plt.xlabel('时段')
plt.ylabel('平均速度（km/h）')
plt.title('平均速度分布曲线图')
plt.tight_layout()#调整布局
plt.show()
print(v['0'].describe())#统计学描述
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#读取数据
time=pd.read_csv('C:/Users/HP/Desktop/XX/zy/time.csv',index_col=0)
distance=pd.read_csv('C:/Users/HP/Desktop/XX/zy/distance.csv',index_col=0)

time.columns=['用时']
distance.columns=['距离']
#统计学描述
print(time.describe())
print(distance.describe())
#画箱线图
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
#定义画图函数
def xiangxiantu(data):
    plt.figure()
    p=data.boxplot(return_type='dict')#画箱线图
    x=p['fliers'][0].get_xdata()#过滤异常值
    y=p['fliers'][0].get_ydata()
    y.sort()#对y进行排序
#作图
xiangxiantu(time)
xiangxiantu(distance)
plt.tight_layout()#自动调整图像界面布局
plt.show()
```

```python
#剔除异常值并对结果进行分析
import pandas as pd
import matplotlib.pyplot as plt
#读取数据，并整理数据
time=pd.read_csv('C:/Users/HP/Desktop/XX/zy/time.csv',index_col=0)
data=pd.read_csv('C:/Users/HP/Desktop/XX/zy/2.csv',index_col=0)
distance=pd.read_csv('C:/Users/HP/Desktop/XX/zy/distance.csv',index_col=0)
hour=pd.read_csv('C:/Users/HP/Desktop/XX/zy/hour.csv',index_col=0,header=None)
data['time']=time.values
data['distance']=distance.values
data['hour']=hour.values
#根据订单用时剔除异常值
IQR=30.400000-13.616667
a=0
b=30.4+1.5*IQR
data=data[(data['time']>a) & (data['time']<b) ]

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
#定义画图函数
def xiangxiantu(data):
    plt.figure()
    p=data.boxplot(return_type='dict')#画箱线图
    x=p['fliers'][0].get_xdata()#过滤异常值
    y=p['fliers'][0].get_ydata()
    y.sort()#对y进行排序
#计算平均速度
v=pd.DataFrame(data['distance']/data['time']*60)
print(v.describe())#对平均速度进行统计学描述
#根据平均速度剔除异常值
vIQR=20.593010-12.077414
vb=20.593010+1.5*vIQR
data['v']=v.values
data=data[v.iloc[:,0]<vb]
data.to_csv('C:/Users/HP/Desktop/XX/zy/3.csv')#存储数据
#重新整理平均速度、订单用时、行驶距离的数据
v=pd.DataFrame(data['v'])
time=pd.DataFrame(data['time'])
distance=pd.DataFrame(data['distance'])
#统计学描述
print(v.describe())
print(time.describe())
print(distance.describe())
#箱线图
xiangxiantu(time)
xiangxiantu(distance)
xiangxiantu(v)
plt.tight_layout()
plt.show()
```
