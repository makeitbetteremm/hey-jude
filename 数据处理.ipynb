```python
#numpy可以创建多种形式的矩阵
import numpy as np
#numpy创建矩阵时可以制定矩阵中元素的类型
print('array,int型(为int32)',np.array([[2,3],[4,5]],dtype=np.int).dtype)
print('array,int64型',np.array([[2,3],[4,5]],dtype=np.int64).dtype)
print('array,float型(为float64)',np.array([[2,3],[4,5]],dtype=np.float).dtype)
print('array,float32型',np.array([[2,3],[4,5]],dtype='float32').dtype)#'float32'另一种引用方式

#常见多种形式的矩阵
print(np.zeros((3,4),dtype=int))#全为0的矩阵
print(np.ones((3,4),dtype=int).reshape(4,3))#创建全为1的矩阵
print(np.empty((3,4)))#创建空矩阵，实际上每个值都为接近于0的数
a=np.empty((3,3))
print('空矩阵第一行',a[0])

print(np.arange(1,10,3).reshape(3,1))#创建连续数组，>=1,<10,步长为3
print(np.linspace(1,10,20))#创建线段型数据，1<=x<=10,且分割成20个数据
```

```python
#numpy基础运算
import numpy as np
#元素的运算
a=np.array([[1,2],[3,4]])
b=np.arange(5,9).reshape(2,2)
print(a,'\n',b)
print('*运算\n',a*b)#元素之间相乘
print('\ 运算\n',a/b)#元素之间相除
print('+运算\n',a+b)#元素之间相加
print('-运算\n',a-b)#元素之间相减
print('比较元素\n',a<3)#对每个元素进行比较，返回 True | False
print('**幂方运算\n',a**2)#每个元素进行幂运算
print('求元素的三角函数\n',10*np.sin(a))#求每个元素的三角函数

#矩阵运算
print('矩阵乘积1\n',np.dot(a,b),a.dot(b))#两种运算方式结果一致,一般np.的表示方法都可以用a.的形式表示

#sum(),max(),min()运算
c=np.random.random((2,3))#创建随机矩阵c
print('c',c)
print('sum',np.sum(c))#计算矩阵全部元素的和
print('max',np.max(c,axis=0),c.max(axis=0))#计算每一列的最大值，也用了两种表示方法
print('min',np.min(c,axis=1))#计算每一行的最小值
```

```python
#numpy基础运算2,利用np.的只能用np.一种方法引用
import numpy as np
a=np.array([[0,1,2],[0,3,4],[0,0,0]])

print('最大值索引',a.argmax())#返回矩阵最大值
print('最小值索引',a.argmin(axis=1))#返回最小值索引，按行，某行相等只返回第一个

print('返回均值',a.mean(axis=1))#返回每行均值
print('返回均值可加权',np.average(a,axis=1,weights=[0,2,2]))#各列权值为【0,2,2】,计算为（a*0+b*2+c*2)/4
print('返回中位数',np.median(a))#排序为0,0,0,0,0,0,1,2,3,4，中位数为0
print('计算累积相加值',a.cumsum())
print('计算累差运算函数',np.diff(a))#每一列前一个数减去后一个数，shape为（3，2）
print('返回非零元素索引',a.nonzero())#返回两个array一维数组，第一个为行索引，第二个为列索引

b=np.arange(12,0,-1).reshape(3,4)
print('b\n',b)
print('对b排序\n',np.sort(b))#默认对行排序，这里不用b.sort()，因为会返回空值。
print('限制矩阵最大最小值\n',b.clip(2,9))#矩阵最小值为2，最大值为9，超过或低于自动赋值为最大或最小值 np.clip(array,array_min,array_max)
print('矩阵转置的两种方法 T\n',b.T,'\ntranspose\n',np.transpose(b))
print(b.flat)
```

```python
#numpy合并
import numpy as np
A=np.array([1,1,1])
B=np.array([2,2,2])
print('A,B',A,B)
C=np.vstack((A,B))
print('C',C)
D=np.hstack((A,B))
print('D',D)

E=A[:,np.newaxis]#在列轴增加新的维度
print('E',E,E.shape)
F=A[np.newaxis,:]#在行轴增加新的维度新的维度
print('F',F,F.shape)
print('A,B np.concatenate',np.concatenate((A,B)))#axis默认为0,维度过低,aixs取值为【0，1）
#concanate可以对多个矩阵进行连接
#设置新矩阵是应为concatenate在一维合并时只有一个轴向可用,如上所示
G=np.ones((1,3))
H=np.ones((1,3))+1
I=np.ones((1,3))+2
print('G,H,I',G,H,I)
J=np.concatenate((G,H,I),axis=1)#行连接(维度1处连接)
print('J',J)
K=np.concatenate((G,H,I),axis=0)#列连接（维度0处连接）
print('K',K)
```

```python
#pandas处理缺失值
import numpy as np
import pandas as pd
ind=pd.date_range('20121212',periods=6)
A=pd.DataFrame(np.random.randn(6,4),index=ind,columns=['A','B','C','D'])
A.iloc[0,1]=np.nan
A.iloc[1,2]=np.nan
#删除具有缺失值的行或列
A.dropna(axis=1,how='any')#aixs=1删除列，how='any'存在即删除，how=’all'，全部空才删除，axis=1应为从行开始寻找，找到则删除列
#将缺失值填充
A.fillna(value=0)#赋值为value
#判断是否有缺失值
A.isnull()#返回True 和 Flase组成的表格

#判断是否由True np.any np.all
np.any(A.isnull(),axis=1)
np.all(A.isnull(),axis=1)
```

```python
#pandas合并concat\append
import numpy as np
import pandas as pd
A=pd.DataFrame(np.ones((3,4))*1,index=range(3),columns=['a','b','c','d'])
B=pd.DataFrame(np.ones((3,4))*2,index=range(1,4),columns=['b','c','d','e'])

#concat
print('join参数默认为outer,ignore_index默认为False,axis默认为0\n',pd.concat((A,B)))#inner取交集outer取并集
print('join=inner  ignore_index=True\n',pd.concat((A,B),join='inner',ignore_index=True))#重置索引
print('axis=1\n',pd.concat((A,B),join='inner',axis=1,ignore_index=True))#行连接,此处会重置列名
print('联结指定索引 axis=0\n',pd.concat((A,B),axis=0,join_axes=[A.columns]))#只联结A列名的列
print('联结指定索引 axis=1\n',pd.concat((A,B),axis=1,join_axes=[B.index]))#只联结B的索引
#append 只能在0维联结，相当于加了新的数据
A.append(B,ignore_index=True)#重置索引
```

```python
#pandas的merge可以合并两组具有共同关键列的数据组
import pandas as pd 
import numpy as np
df1=pd.DataFrame({'key0':['k1','k2','k3','k4'],'key1':['k0','k1','k2','k3'],'r1':range(4),'r2':range(1,5)})
df2=pd.DataFrame({'key0':['k1','k2','k3','k4','k4'],'key1':['k1','k0','k2','k3','k3'],'r1':range(3,8),'r2':range(2,7)})
print('df1\n',df1,'df2\n',df2)
#merge函数的how参数的四种形式inner outer left right
print('innner\n',pd.merge(df1,df2,on=['key0','key1'],how='inner'))#inner取只包括key相同的数据，两个数据组相同key的结果为笛卡儿积
print('outer\n',pd.merge(df1,df2,on=['key0','key1'],how='outer'))#各个数组不同的key用NaN填充
print('right\n',pd.merge(df1,df2,on=['key0','key1'],how='right'))#以右边数组的key列为准同笛卡尔积异空值填充
print('leff\n',pd.merge(df1,df2,on=['key0','key1'],how='left'))#原理同上
#indicator函数可以显示合并后的状态
print('indicator\n',pd.merge(df1,df2,on=['key0','key1'],how='right',indicator=True))#right_only右边有数据，left_only左边，both都有
print("indicator='name'\n",pd.merge(df1,df2,on=['key0','key1'],how='left',indicator='indicator_column'))#对indicator命名后，默认选择True
#当非key列重名时默认加后缀【_x,_y】,suffixes函数可以重命名后缀
print('suffixes 重命名后缀\n',pd.merge(df1,df2,on=['key1','key0'],how='inner',suffixes=['_df1','_df2']))
#根据左右不同列名及索引合并
df3=pd.DataFrame({'key0':['k1','k2','k3','k4'],'r1':range(4),'r2':range(1,5)},index=['k1','k2','k3','k4'])
df4=pd.DataFrame({'key1':['k1','k0','k2','k3','k3'],'r1':range(3,8),'r2':range(2,7)},index=['k1','k2','k3','k4','k5'])
print('df3\n',df3,'df4\n',df4)
print('left_on right_on\n',pd.merge(df1,df2,left_on='key0',right_on='key1',how='outer'))#根据不同名的列合并
print('left_index right_index\n',pd.merge(df1,df2,left_index=True,right_index=True,how='outer'))#根据索引合并
print('left_index right_on\n',pd.merge(df1,df2,left_index=True,right_on='key1',how='outer'))#合并key值在key1列 
```

```python
#matplotlib.pyplot利用bbox修改坐标轴标签的参数
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-3,3,50)
y=x*0.1
#建立基本图像
plt.figure()
plt.plot(x,y,linewidth=10,zorder=1)#经实验坐标轴zorder在2处，ticklabel在1处
plt.ylim(-2,2)
#对坐标轴进行操作
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
#设置坐标轴的标签项
#label是在图像画出前获取的，所以获得的标签值的赋值都为空，若在show()后边再取就会显示赋值
#alpha来设置透明度
for label in ax.get_xticklabels()+ax.get_xticklabels():#取出ax中的x,y坐标轴的标签
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white',edgecolor='none',zorder=2,alpha=0.7))#label的zorder调整到2高于x，y的曲线,bbbox内的值以字典格式读取
plt.show()
```

```python
#matplotlib.pyplot画利用contourf contour画等高线图
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-3,3,256)
y=np.linspace(-3,3,256)
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
X,Y=np.meshgrid(x,y)#meshgrid形成由x,y组成的网格，并返回所有坐标

plt.figure(figsize=(8,5))
plt.contourf(X,Y,f(X,Y),8,alpha=.75,cmap=plt.cm.hot)#levels为8则画出8+2个区域 contourf来fill contour
C=plt.contour(X,Y,f(X,Y),8,colors='k',linewidths=.5)#contour中颜色参数为colors,同理线宽为linewidths
plt.clabel(C,inline=True,fontsize=10)#clabel为contour添加label,inline表示是否在线内显示值

plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#matplotlib.pyplot做3d图
#做3d图需要从mpl_toolkits.mplot3d中导入Axes3D函数，来可视化三维坐标轴
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x=np.arange(-4,4,.25)
y=np.arange(-4,4,.25)
X,Y=np.meshgrid(x,y)
R=np.sqrt(X**2+Y**2)
Z=np.sin(R)

fig=plt.figure()
ax=Axes3D(fig)#在fig中添加三维坐标轴

ax.plot_surface(X,Y,Z,rstride=1,cstride=10,cmap='rainbow')#rsride表示row的跨度，同理cstride表示column跨度

#画3D图的平面等高线图
#zdirection表示从哪个方向压缩3d图形，本处从z轴放行
#offset=-2压缩到点-2处
ax.contourf(X,Y,Z,offset=-2,zdir='z',cmap='rainbow')

ax.set_zlim(-2,1.25)
plt.show()
```

```python
#matplotlib中的animation制作动画
from matplotlib import pyplot
from matplotlib import animation#导入animation库
import numpy as np
x=np.arange(0,2*np.pi,0.01)

fig,ax=pyplot.subplots()
line,=ax.plot(x,np.sin(x))#依旧返回的四list格式
#定义函数返回不同帧的图像
def animate(i):
    line.set_ydata(np.sin(x+i/100))
    return line,
#定义初始导入函数
def init():
    line.set_ydata(np.sin(x))
    return line,
ani=animation.FuncAnimation(fig=fig,func=animate,init_func=init,frames=100,interval=20,blit=True)#frames为帧数，interval为间隔单位ms

pyplot.show()
```

```python
#普通方式爬取莫烦python网址
from bs4 import BeautifulSoup
from urllib.request import urlopen,urljoin
import re
import time
import multiprocessing as mp
count,t1=1,time.time()
base_url='https://morvanzhou.github.io/'
#定义crawl函数获取url的html
def crawl(url):
    response=urlopen(url)
    return response.read().decode('utf-8')
#定义parse函数，解析html
def parse(html):
    soup=BeautifulSoup(html,'lxml')
    title=soup.find('title').get_text().strip()#get_text获取文本内容
    urls=soup.find_all('a',{'href':re.compile(r'^/.+?/$')})
    page_urls=set([urljoin(base_url,url['href']) for url in urls])#urljoin会自动联结href，并删除多余的/
    url=soup.find('meta',{'property':'og:url'})['content']
    return title,page_urls,url
seen=set()
unseen=set([base_url,])

while len(seen)<=20 and len(unseen)!=0:
    htmls=[crawl(url) for url in unseen]
    results=[parse(html) for html in htmls]
    unseen.clear()#clear()清除unseen中的网址
    for title,page_urls,url in results:
        print( count,title,url)
        count+=1
        seen.update(url)
    unseen.update(page_urls-seen)#update()更新set集合
print(time.time()-t1)
```

```python
#多进程爬取用时7s
from bs4 import BeautifulSoup
from urllib.request import urlopen, urljoin
import re
import time
import multiprocessing as mp
count, t1 = 1, time.time()
base_url = 'https://morvanzhou.github.io/'
def crawl(url):
    response = urlopen(url)
    return response.read().decode('utf-8')
def parse(html):
    soup = BeautifulSoup(html, 'lxml')
    title = soup.find('title').get_text().strip()
    urls = soup.find_all('a', {'href': re.compile(r'^/.+?/$')})
    page_urls = set([urljoin(base_url, url['href']) for url in urls])
    url = soup.find('meta', {'property': 'og:url'})['content']
    return title, page_urls, url
seen = set()
unseen = set([base_url, ])
if __name__ == '__main__':
    pool = mp.Pool(4)#创建进程池
    while len(seen) <= 20 and len(unseen) != 0:
        crawl_jobs = [pool.apply_async(crawl, args=(url, )) for url in unseen]#进程池应用异步
        htmls = [j.get() for j in crawl_jobs]#从返回的异步任务中一个个的获取结果
        parse_jobs = [pool.apply_async(parse, args=(html, )) for html in htmls]
        results = [j.get() for j in parse_jobs]
        unseen.clear()
        for title, page_urls, url in results:
            print(count, title, url)
            count += 1
            seen.update(url)
        unseen.update(page_urls-seen)
    print(time.time()-t1)
```
