```python
#拉格朗日插值代码
import pandas as pd #导入数据分析库Pandas
from scipy.interpolate import lagrange #导入拉格朗日插值函数

inputfile ='C:/Users/HP/Desktop/study/python/python data/chapter4/data/catering_sale.xls' #销量数据路径
outputfile ='C:/Users/HP/Desktop/hello.xls' #输出数据路径

data = pd.read_excel(inputfile) #读入数据
data['销量'][(data['销量'] < 400) | (data['销量'] > 5000)] = None #过滤异常值，将其变为空值
#自定义列向量插值函数
#s为列向量，n为被插值的位置，k为取前后的数据个数，默认为5
def ployinterp_column(s, n):
  y = s[list(range(n-5, n)) + list(range(n+1, n+6))] #取数
  y = y[y.notnull()] #剔除空值
  return lagrange(y.index, list(y))(n) #插值并返回插值结果

#逐个元素判断是否需要插值
for i in data.columns:
    for j in range(len(data)):
        if (data[i].isnull())[j]: #j如果为空即插值。
            data.loc[j,i] = ployinterp_column(data[i], j)

data.to_excel(outputfile) #输出结果，写入文件
```

```python
#等频率分布图
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

path ='C:/Users/HP/Desktop/study/python/python data/chapter4/data/discretization_data.xls'
data0=pd.read_excel(path)
data=data0['肝气郁结证型系数'].copy()#拷贝该列成为新的数据集
d1=pd.cut(data,4,labels=range(0,4))#等宽离散化为四个区间
#做d1的图像
plt.figure(1)
for j in range(0,4):
    plt.plot(data[d1==j],[j for i in d1[d1==j]],'o')#d1为对data进行pd.cut后的图像所以可以这样用，用for循环一个点一个点的画出来
plt.ylim(-0.5,3.5)#设置y轴边界
plt.show()

w=[1*i/4 for i in range(5)]#取四分位
w=data.describe(percentiles=w)[4:9]#取data数据描述里边的0%~100%5个分位数
w[0]=w[0]*(1-1e-10)#对最小值进行处理以便分区时定义到最小值
d2=pd.cut(data,w,labels=range(0,4))#等频率离散化
#做d2的图像
plt.figure(2)
d2=pd.DataFrame(d2)
plt.plot(data,d2,'o')
plt.ylim(-0.5,3.5)#设置y轴边界
plt.show()

from sklearn.cluster import KMeans as KM #调取机器学习聚类库中的K聚类
Kmodel=KM(n_clusters=4,n_jobs=4)#n_cluster聚类数，n_jobs为并行数，一般与cpu核数一致
Kmodel.fit(data.values.reshape(len(data),1))#训练模型，reshape将数据转化为列向量
c=pd.DataFrame(Kmodel.cluster_centers_).sort_values(by=0)#取聚类均值，因为均值按随机数据显示的，所以在排序
w=c.rolling(window=2).mean().iloc[1:]#计算相邻均值的均值即聚类边界，并从第二行起取记过，因为第一行为0,index被打乱所以用iloc
v=[0]+list(w[0])+[data.max()]#设置pd.cut的bin，包括0和数据最大值
d3=pd.cut(data,v,labels=range(4))
#做d3的图像
plt.figure(3)
for j in range(0,4):
    plt.plot(data[d3==j],[j for i in d3[d3==j]],'o')#d1为对data进行pd.cut后的图像所以可以这样用，用for循环一个点一个点的画出来
plt.ylim(-0.5,3.5)#设置y轴边界
plt.show()
```

```python
import pandas as pd

path ='C:/Users/HP/Desktop/study/python/python data/chapter5/data/sales_data.xls'
outpath ='C:/Users/HP/Desktop/hello.xls'

data=pd.read_excel(path,index_col=0)
data[data=='好']=1
data[data=='是']=1
data[data=='高']=1
data[data!=1]=0#整理数据


x=data.iloc[:,:3].as_matrix().astype(int)
y=data.iloc[:,3].as_matrix().astype(int)#定义X,Y，定义一个a方便下面引用目录

from keras.models import Sequential
from keras.layers.core import Dense#调用函数

model=Sequential()
model.add(Dense(10,input_shape=(3,),activation='relu'))#input_shape可以写成input_vim=3，该参数定义输入元组的形状为三个元素的向量
model.add(Dense(1,activation='sigmoid'))#这里input_shape默认为第一层的形状
model.compile(loss='binary_crossentropy',optimizer='adam')#定义损失函数和优化器类型
model.fit(x,y,epochs=1000,batch_size=10)#迭代1000次，batch_size为每次训练和梯度更新块的大小一般为32这里数据较少采用10
np=model.predict_classes(x)
np=pd.DataFrame(np,index=range(len(np)))
np.to_excel(outpath)
```

```python
import pandas as pd 
path='C:/Users/HP/Desktop/study/python/python data/chapter5/data/menu_orders.xls'
data=pd.read_excel(path,header=None).as_matrix()
ct=lambda x:pd.Series(1,index=x[pd.notnull(x)])#生成一个一序列，序列索引为x中的非空值
b=map(ct,data)#python3中map返回的值为迭代器（iterator）需要list才能输出
d=pd.DataFrame(list(b))#将1序列转化为Frame格式原Series索引变为列名
d[d.isnull()]=0#建立0-1矩阵，将空值设为0

support=0.2#支持度在所有事务中的占比
confidence=0.5#置信度在上一个项目发生后在发生的概率
ms='--'#设置连接符
result=pd.DataFrame(index=['support','confidence'])
support_series=1.0*d.sum()/len(d)#计算1-项支持度
column=list(support_series[support_series>support].index)#初步筛选支持度
k=0

def connect_string(x,ms):#由已有频繁项集产生新项集
    x=list(map(lambda i:sorted(i.split(ms)),x))#通过map函数将连接好的K-项再次转化为多个元素
    l=len(x[0])#计算第一个项的元素数目即可，也可用k来定义，如l=K+1
    r=[]#产生新的集合
    for i in range(len(x)):
        for j in range(i,len(x)):#两步for循环实现无放回的递次检索
            if x[i][:l-1]==x[j][:l-1] and x[i][l-1]!=x[j][l-1]:#and前保证前k-1项相同，and后保证最后一项不同。无重复找出所有可能
                r.append(x[i][:l-1]+sorted([x[i][l-1],x[j][l-1]]))#沿用相同项加上两个不同项
    return r
while len(column)>1:#频繁项集中项的数量多于两个才可能产生新的频繁项集
    k=k+1
    print('正在进行第%s次检索'%k)
    column=connect_string(column,ms)#定义新的column
    print('新项集数目为%s'%len(column))
    sf=lambda i:d[i].prod(axis=1,numeric_only=True)#计算转化后的各项项行的乘积若为都为1则为1，表示该事务存在该k-项。axis=0为列。
    d_2=pd.DataFrame(list(map(sf,column)),index=[ms.join(i) for i in column]).T#建立连接后的K-项0-1矩阵，.T为转置
    support_series_2=1.0*d_2.sum()/len(d)#计算新一轮的支持度
    column=list(support_series_2[support_series_2>support].index)#筛选出符合支持度的频繁项
    support_series=support_series.append(support_series_2)#整合支持度表
    column2=[]
    for i in column:#遍历k-项集的所有可能规则如A--B--C有三种为A--B~C,A--C~B,B--C~A
        i = i.split(ms)
        for j in range(len(i)):#选取每个j放到最后面产生新的规则
            column2.append(i[:j]+i[j+1:]+i[j:j+1])#i[j:j+1]是list于前面一致，而i[j]为元素需要用list（）转化
    confidence_series=pd.Series(index=[ms.join(i) for i in column2])#由于赋值比较繁琐，索引先单定义
    for i in column2:
        confidence_series[ms.join(i)]=support_series[ms.join(sorted(i))]/support_series[ms.join(sorted(i[:k]))]#计算置信度
    for i in confidence_series[confidence_series>confidence].index:#置信度筛选
        result.loc['confidence',i]=confidence_series[i]
        result.loc['support',i]=support_series[ms.join(sorted(i.split(ms)))]
result=result.T.sort_index(axis=1).sort_values(by='confidence',ascending=False)#度结果进行整理，列索引排序，行数值升序
print('结果为：')
print(result)
```
