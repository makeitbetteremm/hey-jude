```python
import pandas as pd 
path='C:/Users/HP/Desktop/study/python/python data/chapter5/data/menu_orders.xls'
data=pd.read_excel(path,header=None).as_matrix()
ct=lambda x:pd.Series(1,index=x[pd.notnull(x)])#x
b=map(ct,data)#python3mapiteratorlist
d=pd.DataFrame(list(b))#1FrameSeries
d[d.isnull()]=0#0-10

support=0.2#
confidence=0.5#
ms='--'#
result=pd.DataFrame(index=['support','confidence'])
support_series=1.0*d.sum()/len(d)#1-
column=list(support_series[support_series>support].index)#
k=0

def connect_string(x,ms):#
    x=list(map(lambda i:sorted(i.split(ms)),x))#mapK-
    l=len(x[0])#kl=K+1
    r=[]#
    for i in range(len(x)):
        for j in range(i,len(x)):#for
            if x[i][:l-1]==x[j][:l-1] and x[i][l-1]!=x[j][l-1]:#andk-1and
                r.append(x[i][:l-1]+sorted([x[i][l-1],x[j][l-1]]))#
    return r
while len(column)>1:#
    k=k+1
    print('%s'%k)
    column=connect_string(column,ms)#column
    print('%s'%len(column))
    sf=lambda i:d[i].prod(axis=1,numeric_only=True)#11k-axis=0
    d_2=pd.DataFrame(list(map(sf,column)),index=[ms.join(i) for i in column]).T#K-0-1.T
    support_series_2=1.0*d_2.sum()/len(d)#
    column=list(support_series_2[support_series_2>support].index)#
    support_series=support_series.append(support_series_2)#
    column2=[]
    for i in column:#k-A--B--CA--B~C,A--C~B,B--C~A
        i = i.split(ms)
        for j in range(len(i)):#j
            column2.append(i[:j]+i[j+1:]+i[j:j+1])#i[j:j+1]listi[j]list
    confidence_series=pd.Series(index=[ms.join(i) for i in column2])#
    for i in column2:
        confidence_series[ms.join(i)]=support_series[ms.join(sorted(i))]/support_series[ms.join(sorted(i[:k]))]#
    for i in confidence_series[confidence_series>confidence].index:#
        result.loc['confidence',i]=confidence_series[i]
        result.loc['support',i]=support_series[ms.join(sorted(i.split(ms)))]
result=result.T.sort_index(axis=1).sort_values(by='confidence',ascending=False)#
print('')
print(result)
```
